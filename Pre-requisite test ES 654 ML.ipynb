1) How many multiplications and additions in multiplying an nxk and an kxm matrix.

 A) let A be an nxk matrix and B be an kxm matrix.

   The resulting matrix C will be of size nxm.
   
   Now to compute each element in C total number of multiplications = k
   
   Similarly to compute each element in C total number of additions = k-1
   
   Total number of elements in C are n * m.
   
   So total number of multiplications are k * n * m.
   
   And total number of additions are k-1 * n * m.


2) Python code to multiply two matrices.

  (a) Using lists

```python
from random import randint
from datetime import datetime

row = 200
column = 200
x=[]
y=[]
z=[]
for i in range (row):
    a=[]
    for j in range (column):
        a.append(randint(1,10))
    x.append(a)
    
for i in range (row):
    b=[]
    for j in range (column):
        b.append(randint(1,10))
    y.append(b)
    
for i in range (row):
    c=[]
    for j in range (column):
        c.append(randint(0,0))
    z.append(c)

start_time = datetime.now()

for i in range (200):
    for j in range (200):
        for k in range (200):
            z[i][j] += x[i][k] * y[k][j]

end_time = datetime.now()

final_time = end_time - start_time

print (final_time)

```

2) c) Which one is faster and why?

A) As shown above it is clear that matrix multiplication using numpy is faster.

   The diffrence is due to how the data is stored while using lists and how data is stored while using numpy.

   When using lists, the data type is implemented in c as a struct data datatype. In this data type the data is        given a pointer which points to a bunch of pointers to memory addresses which contains the integer data.

   But when numpy is used the data contains one pointer to a memory location which holds an array. The integer data    are stored in continuos memory locations like in an array.

   This defference in the method of storing data not only makes numpy faster to retrieve data but also to store the    computed result, making it faster.

3b) Python code for finding median, the normal method.

```python
L=[1,2,3,4,5,6]

def median(L):
    L.sort()
    if len(L)%2 != 0:
        median = L[int(len(L)/2)]
    
    else:
        median = L[int(len(L)/2)-1] + L[int(len(L)/2)]
        median = median/2
    return median

ans=median(L)
print(ans)
    
```
